```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f1(x,y):
  return np.cos(np.sqrt(3.)*x + y) / (np.pi **2)
def f2(x,y):
  return - np.sin(np.sqrt(3.)*x + y) / (np.pi **2)
```

```python
x = np.linspace(-2.5,2.5,1000)
y = np.linspace(-2.5,2.5,1000)
X, Y = np.meshgrid(x, y)
zs = np.array(f1(np.ravel(X), np.ravel(Y)))
Z1 = zs.reshape(X.shape)
zs = np.array(f2(np.ravel(X), np.ravel(Y)))
Z2 = zs.reshape(X.shape)

fig = plt.figure(figsize=(15,14))

ax = fig.add_subplot(221, projection='3d')
ax.view_init(elev=45.)
ax.set_title("Inverse fourier transform of F1 (3D)")
ax.plot_surface(X,Y,Z1)

ax = fig.add_subplot(222, projection='3d')
ax.view_init(elev=45.)
ax.set_title("Inverse fourier transform of F2 (3D)")
ax.plot_surface(X,Y,Z2,color="darkred")

ax = fig.add_subplot(223)
ax.set_title("Inverse fourier transform of F1 (Contour)")
cf1 = ax.contourf(X,Y,Z1)
fig.colorbar(cf1, ax=ax)

ax = fig.add_subplot(224)
ax.set_title("Inverse fourier transform of F2 (Contour)")
cf2 = ax.contourf(X,Y,Z2)
fig.colorbar(cf2, ax=ax)
```
