    * star graph

    * complete graph
    * complete bipartite graph
    * wheel graph

# Problem #1

```
import networkx as nx
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,6)

g = dict([("star", nx.star_graph(9)),
         ("Petersan", nx.petersen_graph()),
         ("cycle", nx.cycle_graph(10)),
         ("K-10", nx.complete_graph(10)),
         ("K-5,5", nx.complete_bipartite_graph(5, 5)),
         ("wheel", nx.wheel_graph(10)),
      ])

for i, (name,graph) in enumerate(g.items()):
  plt.subplot(2, 6, i+1)
  plt.title(name)
  if name in ["star", "wheel"]:
    nx.draw(graph, node_size=400, node_color='red', with_labels=True, font_weight='bold')
  elif name == "Petersan":
    nx.draw_shell(graph, nlist=[range(5, 10), range(5)], node_size=200, node_color='red', with_labels=True, font_weight='bold')
  elif name == "cycle":
    nx.draw_spring(graph, node_size=400, node_color='red', with_labels=True, font_weight='bold')
  elif name == "K-10":
    nx.draw_circular(graph, node_size=200, node_color='red', with_labels=True, font_weight='bold')
  elif name == "K-5,5":
    pos = nx.bipartite_layout(graph, range(5))
    nx.draw(graph, pos=pos, node_size=400, node_color='red', with_labels=True, font_weight='bold')
  plt.subplot(2, 6, 6 + (i+1))
  plt.plot(nx.degree_histogram(graph))
```
